```python
#3) Faça um programa que leia 10 números inteiros positivos e mostre, no final, a soma dos
# números pares e a soma dos números ímpares.

soma_par = 0
soma_impar = 0

for i in range (1,11):
  n = int(input('Digite um numero: '))
  if n % 2 == 0:
    soma_par = soma_par + n
  elif n % 2 == 1:
    soma_impar = soma_impar + n

print(f'Soma dos números pares é {soma_par}')
print(f'Soma dos números ímpares é {soma_impar}')
```

```python
6#4) Crie um programa que leia 10 números e informe na tela o maior e o menor número lido.
maior = 0
menor = 0

for i  in range (1,11):
  n = int(input('Digite um numero: '))
  if i == 1:
    maior = n
    menor = n
  else:
    if n > maior:
      maior = n
    elif n < menor:
      menor = n

print(f'O maior número foi {maior}')
print(f'O menor número foi {menor}')
```

```python
#6) Digitados dois números (base e expoente ), calcule o resultado  da potência utilizando apenas
#operadores básicos da matemática(soma, subtração, multiplicação ou divisão) e o comando FOR;

b = int(input('Digite a base: '))
e = int(input('Digite o expoente: '))
res = b
if e > 0:
  for i in range (1,e):
    res = res * b
elif e == 0:
  res = 1
3

print(f'{b} elevado a {e} é iqual a {res}')
```

```python
#7) Dado um número n inteiro e positivo, dizemos que n é perfeito se n for igual à soma de
#seus divisores positivos diferentes de n. Construa um programa em python que verifica se um
#dado número é perfeito. Ex: 6 é perfeito, pois 1+2+3 = 6.

n = int(input('Digite um numero: '))
sum = 0
for i in range (1,n):
  if n % i == 0 :
    sum = sum + i

if sum == n:
  print(f'{n} é um numero perfeito')
else:
  print(f'{n} não é um numero perfeito')

```

```python
#8) Escreva um programa em Python que leia um número inteiro positivo e diga se o número informado é primo

n = int(input('Digite um número: '))
cont = 0
if n != 1:
  for i in range (2,n):
    if n % i == 0:
      cont = cont + 1
else:
  cont=cont+1

if cont == 0:
  print(f'O numero {n} é primo')
else:
  print(f'O número {n} não é primo')
```
